```python
class locationHelper:
#     def __init__(self):
#         self._subjec
    def fillBoxLocationInfo(self,tupleTable):
        curBarcodeId = ''
        curSubject = ''
        curType = ''
        curBoxId = ''
        curBoxPos = ''
        newTuple = []
        #tuple should be 64 elements
        for i in range(0,64):
            tmpBarcodeId = tupleTable[0]
            tmpSubject = tmpBarcodeId[3:6]
            tmpBoxId = tupleTable[1]
            curBoxPos = tupleTable[2]
            
            # current barcode is Frozen
            if 'Frozen' in tmpBarcodeId:
                # First met Frozen
                if curType != 'Frozen':
                    curType = 'Frozen'
                    curBarcodeId = tmpBarcodeId
                    curSubject = tmpSubject
                    curBoxId = tmpBoxId
                    curBoxPos = tmpBoxPos
                    
                    newTuple.append([curBarcodeId,curBoxid,curBoxPos])
                # curType is Frozen, but need to double check if subject got changed
                else:
                    # subject different, should start over box id and box position
                    if tmpSubject != curSubject :
                        #curType = 'Frozen'
                        curBarcodeId = tmpBarcodeId
                        curSubject = tmpSubject
                        curBoxId = tmpBoxId
                        curBoxPos = tmpBoxPos
                    
                        newTuple.append([curBarcodeId,curBoxId,curBoxPos])
                    else:
                        # need to update boxid and box position 
                        curBarcodeId = tmpBarcodeId
                        newCurBoxId,newCurBoxPos = self.getBoxIdAndPos(curBoxId,curBoxPos)
                        newTuple.append([curBarcodeId,newCurBoxId,newCurBoxPos])
                        
            else if 'SR' in tmpBarcodeId:
                # First met Frozen
                if curType != 'SR':
                    curType = 'SR'
                    curBarcodeId = tmpBarcodeId
                    curSubject = tmpSubject
                    curBoxId = tmpBoxId
                    curBoxPos = tmpBoxPos
                    
                    newTuple.append([curBarcodeId,curBoxid,curBoxPos])
                # curType is Frozen, but need to double check if subject got changed
                else:
                    # subject different, should start over box id and box position
                    if tmpSubject != curSubject :
                        #curType = 'Frozen'
                        curBarcodeId = tmpBarcodeId
                        curSubject = tmpSubject
                        curBoxId = tmpBoxId
                        curBoxPos = tmpBoxPos
                    
                        newTuple.append([curBarcodeId,curBoxId,curBoxPos])
                    else:
                        # need to update boxid and box position 
                        curBarcodeId = tmpBarcodeId
                        newCurBoxId,newCurBoxPos = self.getBoxIdAndPos(curBoxId,curBoxPos)
                        newTuple.append([curBarcodeId,newCurBoxId,newCurBoxPos])
                        
            else if 'ST' in tmpBarcodeId:
                # First met Frozen
                if curType != 'ST':
                    curType = 'ST'
                    curBarcodeId = tmpBarcodeId
                    curSubject = tmpSubject
                    curBoxId = tmpBoxId
                    curBoxPos = tmpBoxPos
                    
                    newTuple.append([curBarcodeId,curBoxid,curBoxPos])
                # curType is Frozen, but need to double check if subject got changed
                else:
                    # subject different, should start over box id and box position
                    if tmpSubject != curSubject :
                        #curType = 'Frozen'
                        curBarcodeId = tmpBarcodeId
                        curSubject = tmpSubject
                        curBoxId = tmpBoxId
                        curBoxPos = tmpBoxPos
                    
                        newTuple.append([curBarcodeId,curBoxId,curBoxPos])
                    else:
                        # need to update boxid and box position 
                        curBarcodeId = tmpBarcodeId
                        newCurBoxId,newCurBoxPos = self.getBoxIdAndPos(curBoxId,curBoxPos)
                        newTuple.append([curBarcodeId,newCurBoxId,newCurBoxPos])
            else: 
                # no need to set location
                curBarcodeId = tmpBarcodeId
                newTuple.append([curBarcodeId,'',''])
                
    def getBoxIdAndPos(self,tmpCurBoxId, tmpCurBoxPos):
        if int(tmpCurBoxPos) < 81:
            newBoxPos = int(tmpCurBoxPos) + 1
            return tmpCurBoxId,str(newBoxPos)
        else:
            newBoxId = int(tmpCurBoxId) + 1
            newBoxPos = 1
            return str(newBoxId),str(newBoxPos)
        
```
